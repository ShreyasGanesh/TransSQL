```
import json 
import pandas as pd
f = open('/content/drive/MyDrive/685_Project/train_spider.json',) 
data_train = json.load(f) 
```

```
train_int = []
for d in data_train:
  temp = []
  temp.append(d['question'])
  temp.append(d["query"])
  train_int.append(temp)
```

```
train_int_df = pd.DataFrame(train_int, columns=["input_text", "target_text"])
```

```
import logging


from simpletransformers.seq2seq import Seq2SeqModel

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```
model_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": 50,
    "train_batch_size": 128,
    "num_train_epochs": 20,
    "save_eval_checkpoints": False,
    "save_model_every_epoch": False,
    "evaluate_generated_text": True,
    "evaluate_during_training_verbose": True,
    "use_multiprocessing": False,
    "max_length": 25,
    "manual_seed": 4,
    "weight_decay":0.01,
    # "learning_rate":8e-5,
    "warmup_steps":1000,
    "output_dir": '/content/drive/MyDrive/685_Project/outputs_bart_int',
}
model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-base",
    args=model_args,
    use_cuda=True,
)
```

```

model.train_model(train_int_df)
```
